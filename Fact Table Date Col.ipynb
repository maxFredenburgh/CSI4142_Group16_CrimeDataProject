```python
import pandas as pd
import numpy as np
import time


dateDim = pd.read_csv('date_dimension.csv')
denc = pd.read_csv('crimeDen.csv')
#vanc = pd.read_csv('van_crime.csv')

print(denc.head())
#print(vanc.head())
print(dateDim.head())
```

```python
print(len(denc))
```

```python
# Rename dateKey column
dateDim.rename(columns = {'Unnamed: 0':'dateKey'}, inplace = True)
```

```python

vDateKeys = pd.DataFrame(columns=['key']) # VAN surrogate date keys
start = time.perf_counter()
count = 0
for k in range(len(vanc)):
    # retrieve date of crime
    m = vanc['MONTH'][k]
    d = vanc['DAY'][k]
    y = vanc['YEAR'][k]
    
    # match with record in date dimension
    date = dateDim.loc[(dateDim['month'] == m) & (dateDim['dayOfMonth'] == d) & (dateDim['year'] == y)]
    
    j = date['dateKey'].iloc[0] # retrieve key
    vDateKeys.loc[k] = j # record key
    
    # time per 1000 records read
    if k%1000 == 0:
        count += 1000
        end = time.perf_counter()
        print(count, ': Per 1000 records:' + str(end - start) + 's')
        start = time.perf_counter()

# save VAN date keys to csv
vDateKeys.to_csv('vanDateKeys.csv', header=True, encoding='utf-8')
```

```python
dDateKeys = pd.DataFrame(columns=['key']) # denver surrogate date keys
start = time.perf_counter()
count = 0
for k in range(len(denc)):
    rdate = denc['REPORTED_DATE'][k][0:10] # read reported time of crime
    d = rdate.split('/') # format date into month, day, year
    if len(d[2]) > 4:
        d[2] = d[2][0:4]
    # search for matching record in date dimension
    dated = dateDim.loc[(dateDim['month'] == int(d[0])) & (dateDim['dayOfMonth'] == int(d[1])) & (dateDim['year'] == int(d[2]))]

    j = dated['dateKey'].iloc[0] # retrieve key
    dDateKeys.loc[k] = j # record key
    if k%1000 == 0: # time per 1000 records read
        count += 1000
        end = time.perf_counter()
        print(count, ': Per 1000 records:' + str(end - start) + 's')
        start = time.perf_counter()
```

```python
# add isDenver fact
visDenver = pd.Series(0, index=np.arange(len(vDateKeys)))
disDenver = pd.Series(1, index=np.arange(len(dDateKeys)))
vDateKeys['isDenver'] = visDenver
dDateKeys['isDenver'] = disDenver
print(vDateKeys)
print(dDateKeys)
```

```python
dateKeys = pd.concat([vDateKeys, dDateKeys], ignore_index=True)
dateKeys.rename(columns = {'key':'dateKey'}, inplace = True)
dateKeys.to_csv('dateKeys.csv', header=True, encoding='utf-8')
#dateDim.to_csv('dateDimension.csv', header=True, encoding='utf-8')
```
